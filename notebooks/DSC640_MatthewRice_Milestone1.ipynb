```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('/Users/macuser/Downloads/nationaldatabaseofchildcareprices.xlsx')
df.head()
```

```python
tn_df = df[df["State_Name"] == "Tennessee"].copy()
```

```python
tn_df["TotalChildCareCost"] = (
    tn_df["MCInfant"]
    + tn_df["MCToddler"]
    + tn_df["MCPreschool"]
)
```

```python
yearly_totals = (
    tn_df
    .groupby("StudyYear", as_index=False)["TotalChildCareCost"]
    .sum()
)
```

```python
yearly_totals
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 5))

plt.plot(
    yearly_totals["StudyYear"],
    yearly_totals["TotalChildCareCost"],
    marker="o"
)

plt.title("Total Child Care Costs in Tennessee by Year (2008–2018)")
plt.xlabel("Year")
plt.ylabel("Total Cost (USD)")
plt.xticks(range(2008, 2019), rotation=45)

plt.tight_layout()
plt.show()

```

### Summary of Analysis
This analysis uses data from the National Database of Childcare Prices to examine trends in childcare costs in Tennessee between 2008 and 2018. The dataset includes annual county-level estimates for infant, toddler, and preschool childcare costs. For the purposes of this analysis, costs for preschool-aged children (Pre-K) were emphasized, with infant and toddler costs included to contextualize broader childcare affordability trends. The analysis aggregated costs by year to examine how the total cost burden of childcare has changed over time at the state level.

Exploratory data analysis revealed a clear and sustained upward trend in childcare costs across the decade studied. In particular, the total cost associated with preschool care in Tennessee increased from approximately \\$27,000 in 2008 to roughly \\$36,000 in 2018, representing a substantial increase in the financial burden placed on families.

### Findings
The primary finding from this analysis is that childcare costs in Tennessee have risen steadily over time, outpacing what many families can reasonably afford. The increase in preschool-age childcare costs is especially concerning, as this age group represents a critical developmental period for children and a time when many parents are re-entering or fully participating in the workforce.

The data suggests that access to affordable pre-K is not simply a family issue, but a broader economic issue affecting labor participation, household financial stability, and long-term educational outcomes. Rising childcare costs may disproportionately affect lower- and middle-income families, creating barriers to early childhood education and reinforcing existing inequities across communities.

### Assumptions
Several assumptions were made during the analysis. First, childcare costs were aggregated across counties and years to produce statewide trends, assuming this aggregation reasonably reflects the experience of the average Tennessee family. Second, nominal dollar values were used without adjusting for inflation; while this may understate real cost growth, it provides a conservative view of increasing childcare expenses. Finally, it is assumed that the dataset accurately represents prevailing market rates for childcare during the surveyed years.

### Items That Still Need Clarification
Additional clarification would strengthen the analysis, including whether costs are reported as annual full-time equivalents and how rates vary by urban versus rural counties. Further exploration of inflation-adjusted costs and comparisons to median household income would provide deeper insight into affordability. Additionally, understanding current post-2018 trends would help contextualize whether costs have continued to rise at similar or accelerated rates.
